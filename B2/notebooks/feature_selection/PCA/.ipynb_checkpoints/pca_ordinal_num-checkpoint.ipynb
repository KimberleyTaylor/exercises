# Feature Selection
- Principle Component Analysis (PCA)
4. **freq_num**

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
path = "../../../../csv_files/"
```

```python
fulldf_freq_num = pd.read_csv(path + "train/full_train/full_train_combinations/fulldf_freq_num.csv", index_col="Unnamed: 0")
```

```python
testdf_freq_num = pd.read_csv(path + "test/full_test/full_test_combinations/testdf_freq_num.csv", index_col="Unnamed: 0")
```

```python
fulldf_freq_num 
```

```python
X_train = fulldf_freq_num.loc[:,"cat0":"cont12"]
```

```python
X_test = testdf_freq_num.loc[:,"cat0":"cont12"]
```

```python
pca = PCA(n_components=10)
pca.fit(X_train)
```

```python
train_pca_freq_num = pca.transform(X_train)
train_pca_freq_num = pd.DataFrame(train_pca_freq_num)
train_pca_freq_num
```

```python
test_pca_freq_num = pca.transform(X_test)
test_pca_freq_num = pd.DataFrame(test_pca_freq_num)
test_pca_freq_num
```

```python
target = fulldf_freq_num[["id", "target"]]
i_d = pd.DataFrame(testdf_freq_num["id"])
```

```python
train_pca_freq_num = pd.merge(train_pca_freq_num, target, left_index=True, right_index=True)
```

```python
test_pca_freq_num = pd.merge(test_pca_freq_num, i_d, left_index=True, right_index=True)
```

```python
train_pca_freq_num.to_csv("train_pca_freq_num.csv")
```

```python
test_pca_freq_num.to_csv("test_pca_freq_num.csv")
```
