# Bitwise operations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# With numpy create a black canvas of size 800 x 800 
canvas = np.zeros((800,800), np.uint8)
plt.imshow(canvas, cmap="gray")
```

```python
# Make a copy of the canvas and draw a white square of size 250 x 250 on top of the copy of the canvas
square = canvas.copy()
cv2.rectangle(square, (250,250), (500,500), color=(255), thickness=-1)
plt.imshow(square, cmap="gray");
```

```python
# Make a copy of the original canvas and draw a white circle of size 200 x 200 on top of the copy of the canvas
# Make sure the circle is in a different position of the canvas than the square
circle = canvas.copy()
cv2.circle(circle, (450,450), radius=100, color=(255), thickness=-1)
plt.imshow(circle, cmap="gray");
```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise and operation
img_and = cv2.bitwise_and(square, circle)
plt.imshow(img_and, cmap="gray");
```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise or operation
img_or = cv2.bitwise_or(square, circle)
plt.imshow(img_or, cmap="gray");
```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise not operation
img_not = cv2.bitwise_not(square, circle)
plt.imshow(img_not, cmap="gray");
```

```python
# Create a new image and merge both previous images (square and circle) with a bitwise xor operation
img_xor = cv2.bitwise_xor(square, circle)
plt.imshow(img_xor, cmap="gray");
```

```python
ball = cv2.imread("ball.jpg")
ball_copy = ball.copy()
hsv = cv2.cvtColor(ball_copy, cv2.COLOR_BGR2HSV)
plt.imshow(hsv);
```

```python
# Identify a prominent color from the image and filter it out
H, S, V = cv2.split(hsv)
fig, ax = plt.subplots(1,3, figsize=(20, 20))
ax[0].imshow(H, cmap="gray")
ax[1].imshow(S, cmap="gray")
ax[2].imshow(V, cmap="gray");   # See that on the hue image, green is the most prominent color
```

```python
# filter out red
lower_range = (15, 40, 20)
upper_range = (165, 255, 255)

mask = cv2.inRange(hsv, lower_range, upper_range)
red = cv2.bitwise_or(hsv, hsv, mask=mask)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(red, cv2.COLOR_HSV2RGB))
```

```python
mask1 = cv2.bitwise_not(mask)

not_red = cv2.bitwise_and(hsv, hsv, mask=mask1)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(not_red, cv2.COLOR_HSV2RGB));
```

```python
#filter out green

lower_range = (30, 15, 20)
upper_range = (100, 255, 255)

mask = cv2.inRange(hsv, lower_range, upper_range)
green = cv2.bitwise_or(hsv, hsv, mask=mask)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(green, cv2.COLOR_HSV2RGB));
```

```python
mask1 = cv2.bitwise_not(mask)

not_green = cv2.bitwise_and(hsv, hsv, mask=mask1)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(not_green, cv2.COLOR_HSV2RGB));
```
