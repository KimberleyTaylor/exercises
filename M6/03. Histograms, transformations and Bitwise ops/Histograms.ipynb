# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
image1 = cv2.imread("day1.jpg")
image1_copy = image1.copy()
plt.imshow(image1_copy);
```

```python
bgr = image1_copy.copy()

histogram_b = cv2.calcHist([bgr], channels=[0], mask=None, histSize=[255], ranges=[0,255])
histogram_g = cv2.calcHist([bgr], channels=[1], mask=None, histSize=[255], ranges=[0,255])
histogram_r = cv2.calcHist([bgr], channels=[2], mask=None, histSize=[255], ranges=[0,255])

plt.figure(figsize=(10,10))
plt.plot(histogram_b, color='b', label='b')
plt.plot(histogram_g, color='g', label='g')
plt.plot(histogram_r, color='r', label='r')
plt.legend()
plt.title("BGR");
```

```python
hsv = cv2.cvtColor(image1_copy, cv2.COLOR_BGR2HSV)

histogram_h = cv2.calcHist([hsv], channels=[0], mask=None, histSize=[255], ranges=[0,180])
histogram_s = cv2.calcHist([hsv], channels=[1], mask=None, histSize=[255], ranges=[0,255])
histogram_v = cv2.calcHist([hsv], channels=[2], mask=None, histSize=[255], ranges=[0,255])

plt.figure(figsize=(10,10))
plt.plot(histogram_h, color='b', label='h')
plt.plot(histogram_s, color='g', label='s')
plt.plot(histogram_v, color='r', label='v')
plt.legend()
plt.title("HSV");
```
