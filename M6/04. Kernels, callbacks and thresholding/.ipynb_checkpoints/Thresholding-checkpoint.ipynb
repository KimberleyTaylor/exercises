# Computer Vision Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['image.cmap'] = 'gray'
%matplotlib inline
```

```python
# Load an image and apply all types of thresholding that openCV offers (like thresh_binary, thresh_trunc, tozero, etc)
# Display the original image as well as the thresholded images
image = cv2.imread("img/photo1.jpg")
img = image.copy()
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

threshold_types = [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV, cv2.THRESH_TOZERO,
                   cv2.THRESH_TOZERO_INV, cv2.THRESH_TRUNC, cv2.THRESH_MASK]

ret1, thresh1 = cv2.threshold(img, 127, maxval=255, type=threshold_types[0])
ret2, thresh2 = cv2.threshold(img, 127, maxval=255, type=threshold_types[1])
ret3, thresh3 = cv2.threshold(img, 127, maxval=255, type=threshold_types[2])
ret4, thresh4 = cv2.threshold(img, 127, maxval=255, type=threshold_types[3])
ret5, thresh5 = cv2.threshold(img, 127, maxval=255, type=threshold_types[4])
ret6, thresh6 = cv2.threshold(img, 127, maxval=255, type=threshold_types[5])

fig, ax = plt.subplots(3,2, figsize=(15, 15))
ax[0,0].imshow(thresh1[:,::-1])
ax[0,1].imshow(thresh2[:,::-1])
ax[1,0].imshow(thresh3[:,::-1])
ax[1,1].imshow(thresh4[:,::-1])
ax[2,0].imshow(thresh5[:,::-1])
ax[2,1].imshow(thresh6[:,::-1]);
```

```python
# Repeat the above operations with a different image
image = cv2.imread("img/photo2.jpg")
img = image.copy()
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

threshold_types = [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV, cv2.THRESH_TOZERO,
                   cv2.THRESH_TOZERO_INV, cv2.THRESH_TRUNC, cv2.THRESH_MASK]

ret1, thresh1 = cv2.threshold(img, 127, maxval=255, type=threshold_types[0])
ret2, thresh2 = cv2.threshold(img, 127, maxval=255, type=threshold_types[1])
ret3, thresh3 = cv2.threshold(img, 127, maxval=255, type=threshold_types[2])
ret4, thresh4 = cv2.threshold(img, 127, maxval=255, type=threshold_types[3])
ret5, thresh5 = cv2.threshold(img, 127, maxval=255, type=threshold_types[4])
ret6, thresh6 = cv2.threshold(img, 127, maxval=255, type=threshold_types[5])

fig, ax = plt.subplots(3,2, figsize=(15, 15))
ax[0,0].imshow(thresh1[:,::-1])
ax[0,1].imshow(thresh2[:,::-1])
ax[1,0].imshow(thresh3[:,::-1])
ax[1,1].imshow(thresh4[:,::-1])
ax[2,0].imshow(thresh5[:,::-1])
ax[2,1].imshow(thresh6[:,::-1]);
```

```python
# Given this image
text_img = cv2.imread('img/text.png', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (20,15))
plt.imshow(text_img);
```

```python
text_copy = text_img.copy()

ret5, thresh5 = cv2.threshold(text_copy, 127, maxval=255, type=cv2.THRESH_TRUNC)

plt.figure(figsize = (20,15))
plt.imshow(thresh5);
```

```python
# Can you also do the same with this one?
dark_notes = cv2.imread('img/notes.png', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (20,15))
plt.imshow(dark_notes);
```

```python
light_notes = dark_notes.copy()

thresh = cv2.adaptiveThreshold(light_notes, 200, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,31,5)

sharp = np.ones((3,3), np.float32) *-1
sharp[1,1] = 9

thresh = cv2.filter2D(thresh, -1, sharp)


plt.figure(figsize = (20,15))
plt.imshow(thresh);
```
