```python
#!pip install dtreeviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn import tree
from dtreeviz.trees import dtreeviz
from sklearn.model_selection import train_test_split, cross_validate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./datasets/zoo.csv')
```

```python
X = df.drop(["animal_name", "class_type"], axis=1)
Y = df["class_type"]
X.head()
```

```python
Y.head()
```

```python
x = X.values
y = Y.values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=42)
print(X_train[:10])
print(Y_train[:10])
```

```python
clf = DecisionTreeClassifier(splitter='random', max_depth=9, random_state=77)
clf.fit(X_train, Y_train)
```

```python
clf.predict(X_test)
```

```python
print(clf.score(X_train, Y_train))
print(clf.score(X_test, Y_test))
```

```python
print(clf.get_depth())
print(clf.get_n_leaves())
```

```python
columns = []
for col in X.columns: 
    columns.append(col)
```

```python
importance = clf.feature_importances_
importance_df = pd.DataFrame(importance)
importance_df = importance_df.T
importance_df.columns = columns
importance_df
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
fig = plt.figure(figsize=(15,15))
_ = tree.plot_tree(clf, 
                   feature_names=columns,  
                   class_names=df["animal_name"],
                   filled=True)
plt.savefig('tree_zoo.jpg',format='jpg',bbox_inches = "tight")
```
