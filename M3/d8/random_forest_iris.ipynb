```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split, cross_validate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
df = pd.read_csv('../d7/datasets/iris.data', header=None)
df["class"] = iris.target
df.columns = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Class", "Target"]
Y = df["Target"]
X = df.drop(["Class", "Target"], axis=1)
print(X.head())
print(Y.head())
```

```python
columns = []
for col in X.columns: 
    columns.append(col)
```

```python
x = X.values
y = Y.values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=42)
print(X_train[:10])
print(Y_train[:10])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaled_data = MinMaxScaler().fit_transform(X_train)
scaled_data
```

```python
clf = RandomForestClassifier(n_estimators=17, random_state=77)
clf.fit(X_train, Y_train)
```

```python
clf.predict(X_test)
```

```python
print(clf.score(X_train, Y_train))
print(clf.score(X_test, Y_test))
```
